# Superradiance with harmonic radiation

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import sys
sys.path.append('/Users/riverr/Documents/pyfel1d/main/')
from particles import ParticleLoader
from fel import OneDFELSolver, LinearTheory
import time
```

This notebook shows an example of superradiant lasing and the resulting harmonic generation

# Simulation set up 

```python
solver = OneDFELSolver(Lzhat=20, ## undulator length
                       Nzhat=1500, ## number of steps in undulator
                       Lzeta=25,
                       Nzeta=750,
                       time_independent=False)
```

```python
field_rms = 1
field_zeta0 = 5*field_rms
field_amp = 1e-3

solver.a0 = field_amp*np.exp(-(solver.zeta_arr - field_zeta0)**2/(2*field_rms**2))

solver.harmonics = np.array([1, 3, 5])
solver.K0 = 3
```

```python
solver.generate_beam(bdes=0.0)
```

```python
fig, ax = plt.subplots(figsize=(4,3))
ax2 = ax.twinx()

ax2.plot(solver.zeta_arr, np.abs(solver.a0), label='Initial bunching')
ax2.plot(solver.zeta_arr, np.abs(solver.calculate_moments(1,0)), label='Initial bunching')
ax.set_xlabel(r'Beam coordinate $\zeta$')
ax.set_ylabel(r'Beam current profile $\chi$')
ax2.set_ylabel(r'Field and bunching values')
ax2.legend()
```

# Run simulation and look at results

```python
%%time
solver.run_simulation(steps_between_store = 10, ## store the field every 10 steps 
                      verbose=True, ## print status in 10% intervals 
                      store_particles=True)
```

```python
plt.figure(figsize=(4,3))
i = 0 ## harmonic to plot
plt.imshow(1e-12+np.abs(solver.a_store[:,i])**2, aspect='auto', origin='lower', norm=LogNorm(vmin=1e-10), extent=[0, solver.zeta_arr.max(), 0, solver.zhat_store.max()])
plt.xlabel(r'Beam coordinate $\zeta$')
plt.ylabel(r'Undulator coordinate $\hat{z}$')
plt.colorbar(label=f'Power in harmonic {solver.harmonics[i]}')
```

```python
plt.figure(figsize=(4,3))
for i,h in enumerate(solver.harmonics):
    plt.plot(solver.zhat_store, np.mean(np.abs(solver.a_store[:,i])**2, axis=-1), label=f'Harmonic {h}')
plt.xlabel('Undulator position')
plt.ylabel('Average power')
plt.legend()
plt.yscale('log')
plt.ylim(1e-9, 10)
    
```

```python
plt.figure(figsize=(4,3))
nplot = 10

zinds = np.linspace(0, len(solver.zhat_store)-1, nplot).astype(int)

for i in range(nplot):
    zind = zinds[i]
    power = np.abs(solver.a_store[zind,0])**2
    power_norm = power / power.max()
    plt.plot(solver.zeta_arr, i + power_norm, 'k')

plt.plot(solver.zeta_arr, nplot/2*solver.chi, 'r', label=r'Beam profile $\chi$')
plt.xlabel(r'Beam coordinate $\zeta$')
plt.ylabel('Power profiles (arb. units)')
plt.legend()
```

```python
plt.figure(figsize=(4,3))
factors = [1, 250, 3000]
for i,h in enumerate(solver.harmonics):
    plt.plot(solver.zeta_arr, factors[i]*np.abs(solver.a_store[-80,i])**2, label=f'{factors[i]} x Harmonic {h}')
plt.xlabel(r'Beam coordinate $\zeta')
plt.ylabel(r'Power $|a_h|^2$')
plt.legend()
```
